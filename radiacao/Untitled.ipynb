# Radiação script

```python
import pandas as pd
```

```python
df_sujo = pd.read_csv("dataset_irradiacao.csv")
```

```python
df_sujo
```

```python
df = df_sujo[["data","hora","radiacao"]]
```

```python
df
```

```python
dia_count = df["data"].unique()
```

```python
dia_count
```

```python
df_final = pd.DataFrame(columns=['data', 'irradiacao_media'])

for dia in dia_count:
    df_temp=df[df["data"] == dia]
    soma_radiacao = 0
    
    for linha in df_temp.values:
        if linha[1] >= 11 and linha[1] <= 16:
            soma_radiacao += linha[2]
    print(dia, soma_radiacao/6)
    df_final.append({'data':dia,'irradiacao_media':soma_radiacao/6},ignore_index=True)
```
