# Tratamento dos dados de Radiação

Dados sobre radiacao obtidos do INMET - http://www.inmet.gov.br/portal/index.php?r=estacoes/estacoesAutomaticas

Selecionando colunas;

Selecionando entre 11 e 16 horas;

Corrigindo formato da data.

```python
import pandas as pd
import json
```

```python
df_sujo_radiacao = pd.read_csv("dataset_irradiacao.csv")
df_sujo_radiacao.head()
```

```python
df_radiacao = df_sujo_radiacao[["data","hora","radiacao"]]
```

```python
df_radiacao.head()
```

```python
df_radiacao = df_radiacao[(df_radiacao['hora'] >= 11) & (df_radiacao['hora'] <= 16)]
```

```python
df_radiacao['data'] = pd.to_datetime(df_radiacao['data'])
df_radiacao['data'] = df_radiacao['data'].dt.strftime('%d/%m')
```

```python
df_radiacao.head()
```

# Tratamento dos dados de Elevação

Dados sobre elevação solar obtidos em: http://www.heliodon.com.br/download/angulos_solares_DF_BRASILIA.pdf

Criando dataset por data e por hora sobre elevação do sol

```python
df_elevacao = pd.read_csv('./dataset_elevacao.csv', sep=' ')
```

```python
df_elevacao.head()
```

```python
df_elevacao.drop(['declinacao','azm6','azm7','azm8','azm9','azm10','azm11','azm12','azm13','azm14','azm15','azm16','azm17','azm18','alt6','alt7','alt8','alt9','alt10','alt17','alt18'], axis=1, inplace=True)
```

```python
df_elevacao.head()
```

```python
df_elevacao_new = pd.DataFrame(columns=['data','hora','elevacao'])
for col in df_elevacao.values:
    dia = col[0].split('-')[0]
    mes_str = col[0].split('-')[1]
    if(mes_str=='jan'):
        mes='01'
    if(mes_str=='fev'):
        mes='02'
    if(mes_str=='mar'):
        mes='03'
    if(mes_str=='abr'):
        mes='04'
    if(mes_str=='mai'):
        mes='05'
    if(mes_str=='jun'):
        mes='06'
    if(mes_str=='jul'):
        mes='07'
    if(mes_str=='ago'):
        mes='08'
    if(mes_str=='set'):
        mes='09'
    if(mes_str=='out'):
        mes='10'
    if(mes_str=='nov'):
        mes='11'
    if(mes_str=='dez'):
        mes='12'
        
    if int(dia) < 10:
        dia = '0' + str(dia)
        
    data = dia + '/' + mes
    
    hora11 = col[1]
    hora12 = col[2]
    hora13 = col[3]
    hora14 = col[4]
    hora15 = col[5]
    hora16 = col[6]
    df_elevacao_new=df_elevacao_new.append({'data':data,'hora':11,'elevacao':hora11},ignore_index=True)
    df_elevacao_new=df_elevacao_new.append({'data':data,'hora':12,'elevacao':hora12},ignore_index=True)
    df_elevacao_new=df_elevacao_new.append({'data':data,'hora':13,'elevacao':hora13},ignore_index=True)
    df_elevacao_new=df_elevacao_new.append({'data':data,'hora':14,'elevacao':hora14},ignore_index=True)
    df_elevacao_new=df_elevacao_new.append({'data':data,'hora':15,'elevacao':hora15},ignore_index=True)
    df_elevacao_new=df_elevacao_new.append({'data':data,'hora':16,'elevacao':hora16},ignore_index=True)
```

```python
df_elevacao_new['elevacao'] = df_elevacao_new['elevacao'].str.replace(',','.')
```

```python
df_elevacao_new.head()
```

```python
df_elevacao_new.elevacao = df_elevacao_new.elevacao.astype('float64')
```

```python
df_elevacao_new.info()
```

# Dataframe Final

Unindo datasets e gerando json

```python
df_radiacao.head()
```

```python
df_elevacao_new.head()
```

```python
df_final = pd.merge(df_elevacao_new, df_radiacao,  how='left', left_on=['data','hora'], right_on = ['data','hora'])
```

```python
df_final.head(20)
```

```python
data_dict = {}
data_dict['elevacao_radiacao'] = []
for linha in df_final.values:
    data_dict['elevacao_radiacao'].append({
        'data':linha[0],
        'hora':linha[1],
        'elevacao':linha[2],
        'radiacao':linha[3]
    })
    
```

```python
import json
with open("elevacao_radiacao.json", "w") as write_file:
    json.dump(data_dict, write_file, indent=4)
```
